```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("netflix_price.csv")
data
```

```python
a = data["No. of TV Shows"].max()
b = data["No. of Movies"].max()
a,b
```

```python
a1 = data["No. of TV Shows"].min()
b1 = data["No. of Movies"].min()
a1,b1
```

```python
plt.plot(data["No. of TV Shows"])
plt.plot(data["No. of Movies"])
plt.xlabel("Total Countries")
plt.ylabel("Number of Movies/TV Shows")
plt.legend(["Movies", "TV Shows"])
```

```python
sns.pairplot(data=data)
```

```python
sns.histplot(data["Total Library Size"], bins=15, kde=True)
```

```python
plt.plot(data["Cost Per Month - Basic ($)"])
plt.plot(data["Cost Per Month - Standard ($)"])
plt.plot(data["Cost Per Month - Premium ($)"])
plt.legend(["Basic", "Standard", "Premium"])
```

```python
sns.heatmap(data.corr(), cmap='magma', annot=True)
```

```python
top10 = data.nlargest(10, columns=["Total Library Size"])
top10
```

```python
bottom10 = data.nsmallest(10, columns=["Total Library Size"])
bottom10
```

```python
plt.figure(figsize=(14,6))
sns.barplot(top10["Country"], top10["Total Library Size"])
```

```python
plt.figure(figsize=(14,6))
sns.barplot(bottom10["Country"], bottom10["Total Library Size"])
```

```python
sns.barplot(top10["No. of Movies"], top10["Country"])
```

```python
sns.barplot(top10["No. of TV Shows"], top10["Country"])
```

```python
plt.figure(figsize=(14,10))
plt.violinplot(top10["Cost Per Month - Basic ($)"])
plt.violinplot(top10["Cost Per Month - Standard ($)"])
plt.violinplot(top10["Cost Per Month - Premium ($)"])
plt.ylabel("Subscription Cost")
plt.legend()
```
